# Filekezelés pythonban

## Szöveges file beolvasása

### 1. módszer: `open()` + `with`

````python
with open("heroes.txt", "r") as f:
    teljes_file = f.read()

print(f"'{teljes_file}'")
````

#### Hogyan működik?
```python
open("heroes.txt", "r")
```
- megnyitja a paraméterben megadott szöveges file-t (fontos, hogy ez egy mappában legyen a python file-al)
- az `"r"` paraméter pedig azt mutatja, hogy a file olvasásra lett megnyitva
- visszatérési értéke ennek egy úgynevezett "file objektum"

```python
with
```
- ha ezzel kezdődik a filekezelés, akkor a file miután ki lett olvasva automatikusan bezárul

```python
f.read()
```
- kiolvassa a file egész tartalmát és egy szövegként adja vissza

````python
print(type(teljes_file))
````

Látszódik, hogy a `teljes_file` változó típusa `str`, azaz valóban egy string értékként van elmentve ilyen esetben a teljes szöveges file.

### 2. módszer: Sorról sorra olvasás

````python
with open("heroes.txt", "r") as f:
    for sor in f:
        print(f"'{sor.strip()}'")
````

#### Hogyan működik?

- mivel az `f` változóba mentett "file objektum" iterálható, így azon egy `for` ciklussal végighaladva kiolvasható a file sorai egyesével
- az adott sor a `sor` változóba van elmentve
- a `strip()` metódus annyit csinál, hogy egy adott string elején és végén levő üres területeket kitörli a szövegből
- ez azért kell, mert minden sor végén van egy `\n`, ami egy 'ENTER-t' jelöl (új sorba ugrás), így azt ki kell törölni róla

### 3. módszer: Sorok kiolvasása egy listába

````python
with open("heroes.txt", "r") as f:
    sorok = f.readlines()

print(sorok)
````

#### Hogyan működik?

- az `f.readlines()` visszaad egy listát, aminek minden egyes eleme a szöveges file egy sorát tartalmazza
- jól látható a kiírás után, hogy a `sorok` változó minden egyes eleme a szöveges file egy-egy sorát tartalmazza
- az is megfigyelhető, hogy itt a sorok végén ott van a `\n` karakter

Természetesen a sorokat egy listába ki lehet olvasni a 2. módszerrel is, csak abban az esetben létre kell hozni egy listát, amihez hozzá kell fűzni mindig az adott sort:

````python
sorok = []
with open("heroes.txt", "r") as f:
    for sor in f:
        sorok.append(sor)

print(sorok)
````

Látszódik, hogy itt az eredmény ugyanaz lett, mint az `f.readlines()` eredménye
