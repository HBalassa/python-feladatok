## 1. Egyszerű adatszerkezetek

### Típusok

- **Integer (`int`):** Egész számok. Nincs méretkorlátjuk, bármekkorák lehetnek. Pozitív és negatív egyaránt lehet. Pl.: `5`, `-10`
- **Float (`float`):** Úgynevezett lebegőpontos számok, amik lényegében a tizedes törteket jelentik. Ezek is bármekkorák lehetnek, illetve lehetnek pozitív és negatív értékűek is. Pl.: `3.14`, `2.0`, `-0.01`
- **String (`str`):** Ezek úgynevezett karakterláncok, azaz karakterek sorozata, hétköznapi megfogalmazásban szöveges adatok. Idézőjelek (`"`) vagy aposztróf (`'`) közé írjuk. Pl.: `"Hello"`, `'Világ'`
- **Boolean (`bool`):** Logikai érték, ami vagy igaz (`True`), vagy hamis (`False`).

### Létrehozás

```python
x = 10              # int
y = 3.5             # float
nev = "Python"      # string
igaz_e = True       # boolean
```

### Műveletek

#### Számok (`int`, `float`)

- Az egész számokon illetve tizedes számokon az alábbi műveleteket tudjuk végrehajtani: összeadás, kivonás, szorzás, osztás, egész osztás, modulo, hatványozás.
- Fontos megjegyeztni, hogy ha a két tag különböző típusú (az egyik `int` a másik `float`), akkor minden esetben az eredmény `float` típusú lesz!

##### 1. Összeadás (`+`)

```python
print(1 + 2)            # int
print(1.2 + 4.5)        # float
print(3 + 2.3)          # vegyes
```

##### 2. Kivonás (`-`)

```python
print(10 - 6)           # int
print(1.4 - 0.21)       # float
print(3 - 4.5)          # vegyes
```

##### 3. Szorzás (`*`)

```python
print(10 * 2)           # int
print(1.4 * 2.1)        # float
print(5 * 3.2)          # vegyes
```

##### 4. Osztás (`/`): Fontos, hogy ennek eredménye minden esetben `float` típusú szám lesz!

```python
print(10 / 2)           # int
print(3.3 / 1.2)        # float
print(5 / 3.2)          # vegyes
```

##### 5. Egész osztás (`//`): Ez megadja az osztás után az eredmény egész részét, azaz pl. `10 / 3 = 3.3333`, de `10 // 3 = 3`, mert a tizedesvessző utáni részek eltűnnek.

```python
print(10 // 3)          # int
print(11.2 // 3.1)      # float
print(4 // 3.2)         # vegyes
```

##### Modulo (`%`): Ez megadja az osztás maradékát, azaz pl. `14 / 3 = 4.6666`, de `14 % 3 = 2`, mert 14-ben a 3 megvan 4-szer, maradék a 2.

```python
print(14 % 3)           # int
print(12.3 % 2.1)       # float
print(5 % 2.1)          # vegyes
```

##### 7. Hatványozás (`**`): Ahol az első tag mindig az alap, a második tag a hatvány (`alap ** hatvány`)

```python
print(2 ** 3)
print(2.3 ** 3.1)
print(3 ** 1.2)
```

### Összehasonlítások

- Meg tudjuk vizsgálni, hogy egyes értékek között milyen kapcsolat áll fenn
- Két típusa van: az egyenlőség vizsgálata illetve a relációs műveletek
- Az eredménye az összehasonlításoknak mindig egy `bool` típusú érték lesz, azaz vagy igaz vagy hamis

#### 1. Egyenlőség vizsgálata (`==` és `!=`)

- Azt, hogy két érték egyenlő-e minden egyszerű adatszerkezeten meg tudjuk vizsgálni, azaz mind az `int`, `float`, `string` és `bool` értékeken is.
- Különböző típusú értékeken is tudunk vizsgálni egyenlőséget
- Az egyenlőséget a `==` jellel vizsgáljuk, míg azt, hogy nem egyenlő a két érték a `!=` jellel tesszük meg

```python
print(5 == 10)
print(2.3 == 2.3)
print('alma' != 'körte')
print(True == False)

print(5 == 5.0)     # Ez igaz, mert a python tudja, hogy 5.0 ugyanaz mint az 5 csak más típusúak a számok
print(5 == '5')     # Ez viszont hamis, mert az 5 szám nem egyenlő az '5' szöveggel!
```

#### 2. Relációs műveletek (`>`, `<`, `<=` és `>=`)

- Ezeket a műveleteket csak számokon tudjuk elvégezni, azaz csak `int` és `float` típusokon

```python
print(5 > 10)
print(3.4 < 3.41)
print(10 <= 10)
print(5.1 >= 6.7)

print(5 <= 5.0)
```

## 2. Listák

### Mik azok a listák?

- A lista több elem tárolására alkalmas adatszerekezet
- **Rendezett**, azaz az elemeinek a helye a listában számít
- **Módosítható**, azaz lehet az elemeit törölni, módosítani és lehet új elemet hozzáadni
- Bármilyen elemeket képes tárolni és az elemek típusa lehet különböző

### Létrehozása

- Listát létrehozhatunk üresen is, és kezdeti elemekkel is:

```python
 # Egy üres lista létrehozása
ures_lista = []
print(ures_lista)

# Lista létrehozása kezdeti értékekkel
nevek = ['Géza', 'Béla', 'Anna']
print(nevek)

# Lista különböző típusú elemekkel
kutya_adatok = ['Bodri', 3, 12.4, True]     # Itt a kutya neve 'Bodri', aki 3 éves és 12.4 kg-os, illetve hím (ezt jelzi a True)
print(kutya_adatok)
```

### Indexelés

- A listák egyes elemeit elérhetjük a listában levő helyük (indexük) alapján
- A számozás 0-tól indul!
- Így pl. egy számokat tartalmazó lista első eleme: `szamok[0]`.
- Az utolsó elemet pedig megkaphatjuk a `-1` indexszel: `szamok[-1]`

```python
szamok = [1, 32, 10, 4, 67]

# Első elem:
print(szamok[0])

# Utolsó elem:
print(szamok[-1])
```

### Iteráció - lista elemein végighaladni

- a listák elemein végig tudunk haladni több módon is a `for` ciklussal, ezeket mutatom be

#### 1. `for` + `range()`

- Ilyenkor végigmegyünk úgy a listán, hogy egy változóba (általában `i`) tároljuk az index értékét
- a cikluson belül pedig az adott elemet az indexeléssel tudjuk elérni, pl. `szamok[i]`

```python
szamok = [1, 32, 10, 4, 67]

for i in range(len(szamok)):    # Az i változó az adott iterációban mindig az adott index értékét tárolja
    print(i, szamok[i])         # Itt ki is írjuk ezt
```

#### 2. `for elem in lista`

- ilyenkor úgy megyünk végig a listán, hogy csak az egyes elemeket vesszük ki és mentjük mindig egy változóba
- Így az adott elem indexének értékét (azaz, hogy hanyadik az az elem) nem kapjuk meg, viszont jóval könnyebb programozni, ha csak az elemre vagyunk kiváncsiak és az indexre nem

```python
szamok = [1, 32, 10, 4, 67]

for szam in szamok:     # A szam változóba minden iterációban elmentjük a lista soron következő elemének az értékét
    print(szam)         # Majd a cikluson belül tudunk erre hivatkozni, itt kiírjuk
```

#### 3. `for` + `enumerate()`

- Ez az előző esethez hasonló, azonban így nem csak a lista elemét tudjuk megkapni, hanem ki tudjuk menteni az adott elem indexét is
- Ezzel lényegében ugyanazt tudjuk megvalósítani mint az 1. esetben, csak a lista adott elemére nem indexeléssel hivatkozunk, hanem egy változóba van elmentve minden egyes iterációban a soron következő elem

```python
szamok = [1, 32, 10, 4, 67]

for i, szam in enumerate(szamok):   # Ilyenkor az enumerate() függvény segítségével ki tudjuk menteni egy változóba az index értékét (i) illetve a lista adott elemét is (szam)
    print(i, szam)                  # Ezt ki is írjuk
```

- ha létrehozunk egy listát, azon különböző függvényeket meg tudunk hívni, amivel a lista egyes tulajdonságait tudjuk megkapni (pl. hogy mi a legnagyobb eleme, mennyi eleme van...)

```python
pontok = [10, 3, 45, 21, 34]
max_pont = max(pontok)
print(max_pont)

# jegyek = [2, 3, 'jeles', 4, '3']
# max_jegy = max(jegyek) # ez hibát ad, mert nem csak számokból áll a jegyek lista
```

#### 3. Legkisebb elem: `min()`

- a `min(lista)` függvénnyel meg tudjuk kapni egy számokat tartalmazó lista legkisebb elemét

```python
pontok = [10, 3, 45, 21, 34]
min_pont = min(pontok)
print(min_pont)

# jegyek = [2, 3, 'jeles', 4, '3']
# min_jegy = min(jegyek)    # ez hibát ad, mert nem csak számokból áll a jegyek lista
```

#### 3. Elemek összege: `sum()`

#### 4. Elemek sorba rendezése: `sorted()`

- a `sorted(lista)` függvénnyel növekvő sorrendbe tudunk rendezni egy listát, amelyik számokból (`int` vagy `float`) áll, vagy abc sorrendbe egy olyan listát, amely stringekből áll
- Fontos, hogy a szövegből álló listát az angol abc szerint rendezi sorba (szóval amelyik betűk nem tartoznak az angol abc-be azokat a végére sorolja a listának)
- Ha megadunk egy második paramétert is a függvénynek az alábbi módon: `sorted(lista, reverse=True)`, akkor csökkenő sorrendbe rendezi a lista elemeit ha számok, illetve fordított abc sorrendbe ha stringek

```python
nevek_ekezettel = ['Balázs', 'Soma', 'Árpád', 'Ézsau', 'Géza']
nevek_ekezet_nelkul = ['Balazs', 'Soma', 'Arpad', 'Ezsau', 'Geza']

nevek_ekezettel_abc = sorted(nevek_ekezettel)
nevek_ekezet_nelkul_abc = sorted(nevek_ekezet_nelkul)
nevek_ekezet_nelkul_forditott_abc = sorted(nevek_ekezet_nelkul, reverse=True)

print(nevek_ekezettel_abc)                  # Az ékezetekkel kezdődő neveket a lista végére tette
print(nevek_ekezet_nelkul_abc)              # Itt már jól abc sorrendbe tette
print(nevek_ekezet_nelkul_forditott_abc)    # Fordított sorrend


dobasok = [3, 6, 3, 2, 5, 5, 4]

dobasok_novekvo = sorted(dobasok)
dobasok_csokkeno = sorted(dobasok, reverse=True)

print(dobasok_novekvo)
print(dobasok_csokkeno)
```

### Listák metódusai

- a függvényeken kívül a listáknak vannak úgynevezett metódusai is, amik az adott listához tartozó függvényeket jelölik, így ezek közvetlenül a listán fejtik ki hatásukat
- a metódus meghívása mindig úgy történik, hogy a lista után írjuk így: `lista.metodus()`
- a listák metódusai a következőek

#### 1. Elem hozzáadása a lista végére: `.append()`

- a `lista.append(elem)` metódus a lista végére fűzi a paraméterben megadott elemet

```python
nevek = ['Anna', 'Mária', 'József']
nevek.append('Béla')
print(nevek)
```

#### 2. Elem hozzáadása a lista tetszőleges helyére: `.insert()`

- a `lista.insert(index, elem)` metódussal a lista adott indexű helyére tudjuk illeszteni a megadott elemet
- ilyenkor ha már azon a helyen volt egy másik elem, akkor az elemek jobbra tolódnak egy helyet

```python
honapok = ['Március', 'Május', 'Június']
honapok.insert(1, 'Április')
print(honapok)
```

#### 3. Elem törlése: `.remove()`

- a `lista.remove(elem)` metódus az adott elem első megjelenését kitörli a listából
- a törlés után az elem után levő elemeket pedig balra tolja egy hellyel
- ha a paraméterben megadott elem nem tagja a listának, akkor hiba lép fel

```python
jaratok = [133, '101E', 54, 150, 'M4', 54]
jaratok.remove(54)
print(jaratok)      # látszódik, hogy csak az első megjelenését törölte ki az adott elemnek
```

#### 4. Elem kiszedése: `.pop()`

- a `lista.pop(index)` metódussal a lista adott indexű elemét ki tudjuk szedni
- ilyenkor az elem kitörlődik a listából
- ha nincs megadva az `index` paraméter, akkor az utolsó elem lesz kiszedve

```python
vendegek = ['Kiss Béla', 'Nagy János', 'Szabó Attila', 'Deák Árpád']
megerkezett = vendegek.pop(2)
print(megerkezett)
print(vendegek)
```

#### 5. Elem indexének meghatározása: `.index()`

- a `lista.index(elem)` metódussal meg tudjuk határozni, hogy egy adott elem első megjelenése melyik indexen van
- ha a paraméterben megadott elem nincs a listában hibát dob a program

```python
szamok = [1, 2, 40, 20, 54, 80, 20, 12]
szam_index = szamok.index(20)
print(szam_index)       # csak az első megjelenés indexét kapjuk meg
```

#### 6. Elem megjelenésének száma a listába: `.count()`

- a `lista.count(elem)` metódussal meghatározhatjuk, hogy egy adott elem mennyiszer fordul elő a listában

```python
jegyek = [3, 5, 1, 2, 5, 5, 4, 2, 5]
otosok_szama = jegyek.count(5)
print(otosok_szama)
```

#### 7. Lista sorba rendezése: `.sort()`

- a `lista.sort()` metódus a listát növekvő sorba rendezi ha számból állnak az elemei, ha szövegből akkor pedig abc sorrendbe rakja
- fontos, hogy a `sorted()` függvénnyel ellentétben ez a metódus magát a listát megváltoztatja, amin meghívjuk azt
- itt is meg tudunk adni egy paramétert, hogy a lista csökkenő sorrendbe, vagy fordított abc sorrendbe legyen: `lista.sort(reverse=True)`

```python
nevek = ['Adam', 'Eve', 'Jacob', 'Andrew', 'Jonas']

nevek.sort()    # abc sorba rendezés
print(nevek)

nevek.sort(reverse=True)    # fordított abc sorba rendezés
print(nevek)

szamok = [10, 3, 1, 5, 2, 9, 1.2, 3, 4.5]

szamok.sort()   # növekvő sorrendbe
print(szamok)

szamok.sort(reverse=True)   # csökkenő sorrendbe
print(szamok)
```

#### 8. Lista elemeinek sorrendjének megfordítása: `.reverse()`

- a `lista.reverse()` metódussal a lista elemeinek a sorrendjét lehet megfordítani
- itt is annyi különbség van a `reversed()` függvénytől, hogy nem egy új listát hoz ez létre, hanem az adott listát módosítja

```python
ulesrend = ['István', 'Géza', 'Balázs', 'Anna']
ulesrend.reverse()
print(ulesrend)
```

## 3. Listák listái

### Mik azok a listák listái?

- Amint már a listáknál említettem, bármilyen elemet lehet listában tárolni, így egy lista is lehet a lista eleme
- Ha egy lista elemei listák, akkor listák listáiról beszélünk
- Alkalmazása nagyon fontos, így fontos megtudni, hogyan kell programozni ezeket
- Sok dologban ugyanolyan mint a sima lista, így nem mindenre fogok itt kitérni, csak amiben különbözik

### Létrehozása

- Létrehozásuk során minden egyes elem egy lista lesz, azaz:

```python
emberek = [
    ['István', 53, 78.2, True],
    ['Géza', 43, 80.0, True],
    ['Anna', 40, 65.2, False],
    ['Balázs', 30],             # nem kell minden listának ugyanolyannak lennie
]

print(emberek)
```

### Indexelés

- Mivel listák listájáról beszélünk, így egy adott értékre a külső lista és belső lista indexeivel tudunk hivatkozni

```python
emberek = [
    ['István', 53, 78.2, True],
    ['Géza', 43, 80.0, True],
    ['Anna', 40, 65.2, False],
    ['Balázs', 30],
]

adat = emberek[1][2]    # Ez a külső lista 1. elemének a 2. értéke
print(adat)

# Ez máshogy így nézne ki:
ember = emberek[1]      # Ez egy lista
nev = ember[2]          # Ez pedig már egy érték
print(ember)
print(nev)
```

### Iteráció - lista elemein végighaladni

- itt is több lehetőség van, a lényeg az, hogy mindig két `for` ciklust kell alkalmazni
- a külső ciklus mindig a külső lista elemein megy végig (ezek eredménye egy lista)
- a belső ciklus pedig az egyes értékeken megy végig az adott listán

#### 1. `for` + `range()`

- ebben az esetben a külső és belső lista indexeit is megkapjuk a `range()` függvény segítségével

```python
allatok = [
    ['Kefír', 'kutya', 3, 10.2, True],
    ['Cinóber', 'macska', 5, 5.1, True],
    ['Zafír', 'kutya', 3, 10.1, True]
]

for i in range(len(allatok)):               # Először a külső lista indexeit mentjük el az i változóba
    print(f'{i}. lista: {allatok[i]}')
    for j in range(len(allatok[i])):        # Majd pedig a belső lista indexeit a j változóba
        print(i, j, allatok[i][j])          # És ilyenkor tudunk hivatkozni a soron következő elemre az elem[i][j] indexeléssel
```

#### 2. `for lista in listak_listaja`

- ilyenkor először a külső lista egyes elemét elmentjük egy változóba, ami szintén egy lista
- majd utána ezen a listán végigmegyünk és kiszedjük az egyes elemeit

```python
allatok = [
    ['Kefír', 'kutya', 3, 10.2, True],
    ['Cinóber', 'macska', 5, 5.1, True],
    ['Zafír', 'kutya', 3, 10.1, True]
]

for allat in allatok:   # a külső lista adott eleme egy lista
    print(allat)
    for adat in allat:  # azonban ennek a listának az eleme egy érték már
        print(adat)
```

#### 3. `for` + `enumerate()`

- ha nem csak az elemre vagyunk kiváncsiak, hanem az indexre is, akkor itt is lehet alkalmazni az `enumerate()` függvényt

```python
allatok = [
    ['Kefír', 'kutya', 3, 10.2, True],
    ['Cinóber', 'macska', 5, 5.1, True],
    ['Zafír', 'kutya', 3, 10.1, True]
]

for i, allat in enumerate(allatok):     # az index és lista is mentésre kerül
    print(f'{i}. lista: {allat}')
    for j, adat in enumerate(allat):    # illetve a belső index és elem is
        print(i, j, adat)
```

### Függvények és metódusok

- Ezeken is meghívhatóak bizonyos a függvények és metódusok amik a listáknál voltak, annyi külöünbséggel, hogy figyelni kell rá, hogy a külső lista egyes elemei mind listák
- Így a függvények és metódusok nem csak a külső listán, hanem a belső listákon is meghívhatóak
- Pár példa:

```python
diakok = [
    ['Ádám', 12, 'emelt-matek'],
    ['Anna', 11, 'emelt-biosz'],
    ['Fanni', 12, 'emelt-töri']
]

diakok_szama = len(diakok)      # Megkapjuk a külső lista elemeinek a számát
print(diakok_szama)

diakok.append(['Ancsi', 10, 'emelt-info'])  # a külső listához hozzáadunk egy elemet, ami szintén egy lista
print(diakok)

diakok[2].append('emelt-angol')     # A diakok lista 2. eleméhez (ami egy lista) hozzáadunk egy újabb elemet
print(diakok[2])
```

## 4. Szótárak

### Mik azok a szótárak?

- A szótárak olyan adatszerkezetek, amely kulcs-érték párokat tárol
- A kulcsnak egyedinek kell lennie és nem lehet megváltoztatni
- Egy kulcshoz mindig csak egy érték tartozik, azonban az érték lehet pl. egy lista is, amibe tudunk több értéket is eltárolni

### Létrehozásuk

- lehet üresen vagy kezdeti értékekkel is létrehozni:

```python
diak_ures = {}      # üres szótár

diak = {
    'név': 'Balázs',                            # Mindig kell egy : a kulcs és érték közé létrehozáskor
    'kor': 18,
    'évfolyam': 12,
    'barátok': ['Béla', 'Géza', 'István']       # Lehet lista is az érték
}

print(diak)
```

### Módosítás, elérés

- a szótárak elemeit nem lehet indexeléssel elérni, hanem a kulcs megadásával kell az egyes értékekre hivatkozni:

```python
kutya = {
    'név': 'Bodri',
    'kor': 3,
    'súly': 4.5,
    'nem': True
}

kutya_neve = kutya['név']       # így tudunk elérni egy adott kulcshoz tartozó értéket
print(kutya_neve)

kutya['kor'] += 1               # a kulccsal hivatkozva tudunk változtatni az értéken
print(kutya['kor'])

kutya['szín'] = 'fekete'        # ha még nincs ilyen kulcs, akkor létrehoz egy új kulcsot a szótárban az adott értékkel
print(kutya)
```

### Iteráció - elemeken végighaladni

- bár a szótárakon nem lehet indexeléssel végigmenni, de a kulcs-érték párokon végig lehet haladni különböző módszerekkel:

#### 1. Kulcsokon végighaladás

- ha csak a kulcsokat szeretnénk megkapni egyesével, akkor azt az alábbi módon lehet megtenni:

```python
ember = {
    'név': 'Nagy István',
    'kor': 54,
    'súly': 83.4,
    'magasság': 185,
    'házas-e': True,
    'gyerekek': ['Nagy Ottó', 'Nagy Anna']
}

for kulcs in ember.keys():
    print(kulcs)        # így a kulcsokon tudunk végigmenni
```

#### 2. Értékeken végighaladás

- ha csak az értékeket szeretnénk egyesével megkapni azt így lehet:

```python
ember = {
    'név': 'Nagy István',
    'kor': 54,
    'súly': 83.4,
    'magasság': 185,
    'házas-e': True,
    'gyerekek': ['Nagy Ottó', 'Nagy Anna']
}

for ertek in ember.values():        # Így pedig csak az értékeken tudunk végigmenni
    print(ertek)
```

#### 3. Kulcs-érték pároson végigmenni

- azonban azt is megtehetjük, hogy mind a kulcsot és az értéket szeretnénk megkapni egyesével:

```python
ember = {
    'név': 'Nagy István',
    'kor': 54,
    'súly': 83.4,
    'magasság': 185,
    'házas-e': True,
    'gyerekek': ['Nagy Ottó', 'Nagy Anna']
}

for kulcs, ertek in ember.items():      # Így pedig megkapjuk mind a kulcsot és az értéket is
    print(kulcs, ertek)
```

### Szótárak metódusai

- a szótárakon is vannak metódusok, ezek közül a legfontosabbak:

#### 1. Kulcsok listája: `.keys()`

- a `szotar.keys()` metódussal megkapjuk a kulcsokat egy listához hasonló objektumban

```python
auto = {
    'márka': 'Opel',
    'modell': 'Combo',
    'évjárat': 2019,
    'futás_km': 54785,
    'utasok_száma': 7
}

kulcsok = auto.keys()
print(kulcsok)
```

#### 2. Értékek listája: `.values()`

- a `szotar.values()` metódussal le tudjuk kérni az értékeket egy listához hasonló objektumban

```python
auto = {
    'márka': 'Opel',
    'modell': 'Combo',
    'évjárat': 2019,
    'futás_km': 54785,
    'utasok_száma': 7
}

ertekek = auto.values()
print(ertekek)
```

#### 3. Kulcs-érték párok listája: `.items()`

- a `szotar.items()` metódussal megkapjuk a kulcs-érték párosokat egy listák listájához hasonló objektumban

```python
auto = {
    'márka': 'Opel',
    'modell': 'Combo',
    'évjárat': 2019,
    'futás_km': 54785,
    'utasok_száma': 7
}

elemek = auto.items()
print(elemek)
```

## 5. Halmazok

### Mik azok a halmazok?

- olyan több adat tárolására alkalmas adatszerkezet, amelynek minden eleme egyedi
- azaz lényegében olyasmi mint egy lista, csak minden eleme különböző (nem lehetnek duplikátumok)
- nem rendezett ez sem, azaz nem lehet indexeléssel hvatkozni az elemekre

### Létrehozásuk

- lehet üresen és kezdeti értékekkel is létrehozni, illetve lehet listából is létrehozni

```python
ures_halmaz = set()
print(ures_halmaz)

halmaz = {1, 4, 10, 43, 23}
print(halmaz)

# Legfontosabb létrehozása ha listából szeretnénk megtudni mik azok az elemek amik a listában vannak
szamok = [2, 3, 5, 5, 1, 2, 5, 3, 4]
egyedi_szamok_halmaza = set(szamok)
print(egyedi_szamok_halmaza)

# majd ezt vissza is tudjuk alakítani listává
egyedi_szamok_lista = list(egyedi_szamok_halmaza)
print(egyedi_szamok_lista)
```

### Iteráció - elemeken végighaladni

- Mivel nem rendezettek, így nem indexelhetőek, de végig lehet haladni a `for` ciklussal az egyes elemeken:

```python
napok = {'02.12', '03.31', '05.14', '05.23', '06.11', '12.29'}

for nap in napok:
    print(nap)
```

### Metódusok

- a halmazok legfontosabb metódusai:

#### 1. Új elem hozzáadása: `.add()`

- a `halmaz.add(elem)` hozzáad egy elemet a halmazhoz, ha az még nincs benne
- ha már benne van, akkor nem csinál semmit

```python
lepesek = {5, 7, 12, 13, 17, 18, 22, 26}
lepesek.add(26)     # mivel már benne van, így nem adja hozzá
lepesek.add(30)
print(lepesek)
```

#### 2. Meglévő elem kitörlése: `.discard()`

- a `halmaz.discard(elem)` kitörli az adott elemet, ha tagja a halmaznak, ha nem, akkor nem történik semmi

```python
lepesek = {5, 7, 12, 13, 17, 18, 22, 26}
lepesek.discard(26)
lepesek.discard(28)     # nem történik semmi
print(lepesek)
```

### Halmazműveletek

- a halmazokon a matematikában ismert halmazműveleteket ugyanúgy el tudjuk végezni
- ezek a következőek (legfontosabbak): unió, metszet, különbség, szimmetrikus differencia

#### 1. Unió (`|`)

- az unió műveletével egy olyan halmazt kapunk, mely az eredeti két halmaz minden elemét tartalmazza

```python
a = {1, 2, 3}
b = {3, 4, 5}
unio = a | b
print(unio)
```

#### 2. Metszet (`&`)

- a metszet műveletével olyan halmazt kapunk, mely az eredeti két halmaz közös elemeit tartalmazza

```python
a = {1, 2, 3}
b = {3, 4, 5}
metszet = a & b
print(metszet)
```

#### 3. Különbség (`-`)

- a különbség műveletével olyan halmazt kapunk, mely tartalmazza az első halmaz elemeit, de a második halmaz elemeit nem

```python
a = {1, 2, 3}
b = {3, 4, 5}
kulonbseg_a = a - b     # ez minden olyan elem, amelyik a-ban van, de b-ben nincs
kulonbseg_b = b - a     # ez viszont minden olyan elem, amelyik b-ben van, de a-ban nincs
print(kulonbseg_a)
print(kulonbseg_b)
```

#### 4. Szimmetrikus különbség (`^`)

- a szimmetrikus különbség műveletével olyan halmazt kapunk, amely elemei a két halmaz közül pontosan az egyikben szerepelnek, de a metszetben nem
- azaz egy olyan "unio", amibe a metszet elemeit nem vesszük bele

```python
a = {1, 2, 3}
b = {3, 4, 5}
szimmetrikus_kulonbseg = a ^ b
print(szimmetrikus_kulonbseg)
```
